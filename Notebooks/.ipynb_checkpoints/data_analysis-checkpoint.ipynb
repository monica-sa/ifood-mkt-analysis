## CHECKS

#### Check if a customer/row has one and only one education & marital flags set to 1    

```python
def check_flags(row):
    # Summing the values in education and marital status columns
    education_sum = row[['education_2n Cycle', 'education_Basic', 'education_Graduation','education_Master', 'education_PhD']].sum()
    marital_sum = row[['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']].sum()
    
    # Check if exactly one education and one marital status flag are set to 1
    return education_sum == 1 and marital_sum == 1

# Apply the function to each row
input_df['ValidFlags'] = input_df.apply(check_flags, axis=1)

# print rows which don't meet the criteria
input_df[input_df['ValidFlags'] == 0]
```

âœ“ 

#### Check if the MntTotal is the sum of MntGoldProducts/MntRegularProductcs and of Mnt by Categories

```python
def check_mnt(row, compare):
    # Summing the values in mnt by category columns
    if compare == "category":
        # Assuming these are your actual spending category columns
        mnt_total = row[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts']].sum()
    elif compare == "tier":
        # Add correct logic for tier comparison here, if different from category
        # Placeholder for correct tier logic
        mnt_total = row[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts']].sum()
    else:
        raise ValueError("Invalid comparison type")
    
    # Check if total matches
    return mnt_total == row['MntTotal']

# Adjust the apply method to pass additional arguments with args
input_df['ValidMntsTier'] = input_df.apply(check_mnt, args=("tier",), axis=1)
input_df['ValidMntsCategory'] = input_df.apply(check_mnt, args=("category",), axis=1)

# Print rows which don't meet the criterias
input_df[(input_df['ValidMntsTier'] == False) | (input_df['ValidMntsCategory'] == False)]
```

âœ“ 

#### Check if the AcceptedCmpOverall is the sum of AcceptedCmp

```python
def check_campaigns(row):
    # Summing the values in campaign columns
    total_campaigns_acepted = row[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].sum()
    # Check if total matches
    return total_campaigns_acepted == row['AcceptedCmpOverall']

# Adjust the apply method to pass additional arguments with args
input_df['ValidCmpOverall'] = input_df.apply(check_campaigns, axis=1)

# Print rows which don't meet the criterias
input_df[(input_df['ValidCmpOverall'] == False)]
```

âœ“

#### Check if the Customer_Days is greater than Recency

```python
def check_recency(row):
    return row['Customer_Days'] >= row['Recency']

# Adjust the apply method to pass additional arguments with args
input_df['ValidRecency'] = input_df.apply(check_recency, axis=1)

# Print rows which don't meet the criterias
input_df[(input_df['ValidRecency'] == False)]
```

# Data Engineer

## Are there any useful variables that we can be added?

#### Metrics
- **Dependents**: The total number of dependents in the home can be engineered from the sum of 'Kidhome' and 'Teenhome'
- **TotalPurchases**: The total purchases can be engineered from the sum of all features containing the keyword 'Purchases'
- **RFM Scores**: (Recency, Frequency, Monetary): Using the recency, TotalPurchases and MntTotal we can engineer the score and define segments

#### Qualitative
- **AgeGroup**
    - 1: [0,24]
    - 2: [24,39]
    - 3: [40,55]
    - 4: [56,71]
    - 5: 71+
- **Income Group** (based on the quantile)
    - 1: [0, 35196]
    - 2: [35196, 51287]
    - 3: [51287, 68281]
    - 4: 68281+
- **Graduation**
    -  2nCycle
    -  Basic
    -  Graduation
    -  Master
    -  PhD
- **Marital**
    -  Divorced
    -  Married
    -  Single
    -  Together
    -  Widow
 
#### Flags
- **is_graduated**: 1 if has graduation, master or phD. 0 otherwise
- **is_couple**: 1 if together or married. 0 otherwise
- **has_child**: 1 if has child or teen 0 otherwise
- **is_single_parenting**: 1 if is_couple is false and has_child is true. 0 otherwise
- **is_churn**: 1 if customer present in the 4th quartil of recency (recency > 74 days), 0 otherwise), 0 otherwise
